## Margin Strat

```python
class backtest():
    def __init__(self,data, fund = 10000, cursor_pos = 30, verbose = False):
        if fund <= 0:
            raise ValueError('Invalid initial fund value: {}'.format(fund))
        self.dat  = data
        self.dat.index = range(self.dat.shape[0])
        
        self.invest_ratio = 0.5  #invest percentage of the account balance
        self.max_loss = 0.2      #stop
        self.max_gain = 0.15     #stop
        self.verbose = verbose
        
        tmp = np.empty(shape = (data.shape[0],5))
        tmp.fill(np.NaN)
        self.account_hist = pd.DataFrame(tmp, columns = ['datetime','signal','position','cost','balance'])
        
        self.fund = fund         #account balance
        self.initial_fund = fund
        
        self.position = 0.0
        self.cost = np.NaN       #cost of position
        #self.pnl = 0.0
        self.action = 'wait'     #action for this period, decided by last time period
        self.cursor = cursor_pos #position in dat
        
    def backtest(self):
        start_time = datetime.datetime.now()
        
        print('Backtesting on ', self.dat.shape[0] - self.cursor, 'time steps')
        self.trades_counter = 0
        
        while self.cursor < self.dat.shape[0]:
            idx = self.cursor
            pos_changed = False
            close_price = self.dat.loc[idx,'close']
            
            if self.position == 0:
                if self.action == 'wait': #Find signal
                    if self._signal() == True:
                        self.action = 'long'
                elif self.action == 'long':
                    print('-  LONG  -')
                    pos_changed = True
                    self.trades_counter += 1
                    self.cost = close_price
                    self.position = self.fund*self.invest_ratio/self.cost
                    self.fund = self.fund - self.position * self.cost
                    self.action = 'wait'
            else:
                if self.action == 'sell':
                    print('-  SELL  -')
                    pos_changed = True
                    self.trades_counter += 1
                    self.fund = self.fund + self.position * close_price
                    self.cost = close_price
                    self.position = 0
                    self.action = 'wait'
                else:
                    ret = (- close_price + self.cost) / self.cost
                    if ret < -self.max_loss or ret > self.max_gain:
                        self.action = 'sell'
                        
            self._log(pos_changed)
            self.cursor += 1
        
        print("Backtest finished in {}, traded {} times\n initial fund: {}, end fund: {}\n return: {} ".format(
             datetime.datetime.now()-start_time,
             self.trades_counter,
             self.initial_fund,
             self.fund - self.position * close_price,
             (self.fund - self.position * close_price)/self.initial_fund - 1))
    
    def _signal(self):
        idx = self.cursor
        if self.dat.loc[idx, 'ma_5'] < self.dat.loc[idx, 'ma_30'] and \
                    self.dat.loc[idx-1, 'ma_5'] > self.dat.loc[idx-1, 'ma_30']:
                        return True
        else:
            return False
    
    def _log(self, pos_changed):
        idx = self.cursor
        close_price = self.dat.loc[idx,'close']
        self.account_hist.loc[idx,:] = [self.dat.loc[idx,'datetime'],self._signal(),self.position,self.cost,self.fund]
        if self.verbose == True and pos_changed == True:
            print(' | [{}] | position: {:<7} | cost: {:<12} |\n | balance: {:<12} | PnL: {:<12}'.format(
                self.dat.loc[idx,'datetime'], 
                round(self.position,3),
                round(self.cost,3), 
                round(self.fund,3),
                round(self.position*close_price + self.fund - self.initial_fund)))
```

```python
for i in [1,5,10,15,30]:
    hist_price['ma_'+str(i)] = hist_price[['Total Amount Active Swaps (BTC)']].rolling(window = i).mean()
```

```python
bt = backtest(hist_price, verbose = True)
```

```python
bt.backtest()
```
